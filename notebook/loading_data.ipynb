```python
from kiwipiepy import Kiwi
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_name = "tabularisai/multilingual-sentiment-analysis"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

from kiwipiepy import Kiwi

kiwi = Kiwi(model_type='knlm', typos='basic_with_continual_and_lengthening')

def split_review_into_sentences(text):
    rlt = []
    sents = kiwi.split_into_sents(text, normalize_coda=True)
    for s in sents:
        if '\n' in s.text:
            s_splited = s.text.split('\n')
            for s_tmp in s_splited:
                rlt.append(s_tmp.strip().replace('.', ''))
        else:
            rlt.append(s.text.strip().replace('.', ''))
    return rlt


def predict_sentiment(texts):
    inputs = tokenizer(texts, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    probabilities = torch.nn.functional.softmax(outputs.logits, dim=-1)
    sentiment_map = {0: "Very Negative", 1: "Negative", 2: "Neutral", 3: "Positive", 4: "Very Positive"}
    return [sentiment_map[p] for p in torch.argmax(probabilities, dim=-1).tolist()]



def process_reviews(review_table):
    sentence_nodes = []  # 전체 문장 노드
    for i, row in review_table.iterrows():
        review_id = row['review_id']
        reviewer_id = row['reviewer_id']
        diner_id = row['diner_idx']
        text = row['reviewer_review']
        if not text:
            continue
        sentences = split_review_into_sentences(text)

        for s in sentences:
            if len(s) < 10:
                continue
            sentence_nodes.append({
                'review_id': review_id,
                'reviewer_id': reviewer_id,
                'diner_id': diner_id,
                'sentence': s,
                'sentiment': predict_sentiment(s)
            })
    return sentence_nodes
```

```python
import re

def clean_menu_name(menu):
    stopwords = ['1인식사','1인분', '1인', '1인세트', '1인당','메뉴', '추가', '등급이상', '등급', '단', '년', 'kg', 'mm', '일 ', 'g', '만 ', '인 ', '인분', '당 ']
    for stop in stopwords:
        menu = menu.replace(stop, "")
    menu = re.sub(r"[^\uAC00-\uD7A3\s]", "", menu)  # 한글/영어/숫자 외 제거
    menu = re.sub(r"\s+", " ", menu).strip()  # 공백 정리
    if menu == '':
        return None
    return menu
```

```python
from kiwi import Kiwi
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModel, AutoModelForSequenceClassification
import torch.nn.functional as F

# 모델 초기화
kiwi = Kiwi(model_type='knlm', typos='basic_with_continual_and_lengthening')
koe5_tok = AutoTokenizer.from_pretrained("Beomi/KoE5-base-v1")
koe5_model = AutoModel.from_pretrained("Beomi/KoE5-base-v1")
sent_tok = AutoTokenizer.from_pretrained("beomi/KcELECTRA-base")
sent_model = AutoModelForSequenceClassification.from_pretrained("beomi/KcELECTRA-base")


# 조건 1: 최소 문장 길이 및 내용성
def is_meaningful_sentence(tokens, min_length=3):
    content_tags = {'NNG', 'NNP', 'VV', 'VA', 'MAG'}
    if len(tokens) < min_length:
        return False
    return any(tok.tag in content_tags for tok in tokens)

# 조건 2: 메뉴명이 포함되어 있는지
def mentions_menu(text, menu_names):
    return any(menu in text for menu in menu_names)

# 문장 임베딩 (384-d)
def get_sentence_embedding(sentence):
    inputs = koe5_tok(sentence, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        output = koe5_model(**inputs)
    embedding = output.last_hidden_state.mean(dim=1)  # [1, 384]
    return embedding.squeeze()  # [384]

# 감성 점수 [-1, 1] → [0, 1] 정규화
def get_sentiment_score(sentence):
    inputs = sent_tok(sentence, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        logits = sent_model(**inputs).logits
        probs = F.softmax(logits, dim=1)
        score = probs[0][2].item() - probs[0][0].item()  # positive - negative
        return (score + 1) / 2

# 전체 파이프라인 함수
def process_review_text(text, menu_names):
    results = []
    sents = kiwi.split_into_sents(text, return_tokens=True, normalize_coda=True)
    for sent in sents:
        if '\n' in sent.text:
            sent_splited = sent.text.split('\n')
            for sent_tmp in sent_splited:
                if not is_meaningful_sentence(sent_tmp.tokens):
                    continue
                if not mentions_menu(sent_tmp.text, menu_names):
                    continue
                emb = get_sentence_embedding(sent_tmp.text)
        if not is_meaningful_sentence(sent.tokens):
            continue
        if not mentions_menu(sent.text, menu_names):
            continue
        emb = get_sentence_embedding(sent.text)
        sentiment = get_sentiment_score(sent.text)
        combined = torch.cat([emb, torch.tensor([sentiment])])  # [385-d]
        results.append((sent.text, combined))
    return results
```

```python
import pandas as pd

# 타임포인트 정의
train_time_point = pd.to_datetime("2024-09-01")
val_time_point = pd.to_datetime("2024-12-01")
test_time_point = pd.to_datetime("2025-01-01")
end_time_point = pd.to_datetime("2025-02-01")

# 날짜 컬럼을 datetime으로 변환
review['reviewer_review_date'] = pd.to_datetime(review['reviewer_review_date'])

# 분석 대상 reviewer만 필터링
subset_df = review.copy()

# 평균 글자 수 계산 함수
def avg_length(series):
    return series.dropna().apply(len).mean()

# 그룹별 집계
aggregated_df = subset_df.groupby('reviewer_id').agg(
    total_reviews=('review_id', 'count'),
    non_null_reviews=('reviewer_review', lambda x: x.notnull().sum()),
    avg_chars_per_review=('reviewer_review', avg_length),
    train_val_reviews=('reviewer_review_date', lambda x: ((x >= train_time_point) & (x < val_time_point)).sum()),
    train_test_reviews=('reviewer_review_date', lambda x: ((x >= train_time_point) & (x < test_time_point)).sum()),
    val_reviews=('reviewer_review_date', lambda x: ((x >= val_time_point) & (x < end_time_point)).sum()),
    test_reviews=('reviewer_review_date', lambda x: ((x >= test_time_point) & (x < end_time_point)).sum())
).reset_index()

# 결과 확인
filtered_df = aggregated_df[(aggregated_df[['train_val_reviews', 'train_test_reviews', 'val_reviews', 'test_reviews']] != 0).all(axis=1)]

qualitative_evaluation_reviewer_id = filtered_df[(filtered_df['avg_chars_per_review']>=filtered_df['avg_chars_per_review'].mean()) & (filtered_df['test_reviews']>=10)].sort_values(by='train_val_reviews', ascending=False).head(10)['reviewer_id'].tolist()
```
