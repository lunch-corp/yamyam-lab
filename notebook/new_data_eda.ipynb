```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
review = pd.read_csv("/Users/billy.s/Downloads/review.csv")

review["date"] = review["date"].map(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
review["date_month"] = review["date"].map(lambda x: x.replace(day=1, hour=0, minute=0, second=0))
```

```python
review.shape
```

```python
num_diners = len(review['diner_idx'].unique())
num_reviewers = len(review['reviewer_id'].unique())

print(f"리뷰를 남긴 unique 식당 수: {num_diners}")
print(f"리뷰를 남긴 unique 리뷰어 수: {num_reviewers}")
```

```python
review.date_month.value_counts().sort_index().plot()
plt.xticks(rotation=20)
plt.show()
```

```python
review[lambda x: x["date_month"] >= datetime(year=2024, month=1, day=1)].date_month.value_counts().sort_index().plot()
plt.xticks(rotation=20)
plt.show()
```

* 25년 12월 리뷰의 수가 현격하게 낮음
* 이전 크롤링 데이터도 직전달의 리뷰수가 적었음
* 시스템 상으로 늦게 반영되는 것 같음

```python
review[lambda x: x["date_month"] >= datetime(year=2025, month=6, day=1)].value_counts("date_month")
```

* 데이터는 여전히 skewed 되어 있음.
* 리뷰를 1개쓴 리뷰어의 수가 전체 63%, 2개 쓴 리뷰어의 수가 14%임
* 전체의 80% 정도가 리뷰를 3개 이하 작성함

```python
review["reviewer_id"].value_counts().reset_index()["count"].value_counts().head(20) / num_reviewers
```

* 식당별 리뷰수 분포는 그나마 좀 나은 편임
* 리뷰를 1개 받은 식당이 전체의 9% 정도임

```python
review["diner_idx"].value_counts().reset_index()["count"].value_counts().head(20) / num_diners
```
