```python
import pandas as pd

pd.set_option("display.max_columns", 30)
import os
import glob
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid")

DATA_PATH = os.path.join(os.path.abspath(""), "../data")
# Load data
review_data_paths = glob.glob(os.path.join(DATA_PATH, "review", "*.csv"))
```

```python
diner = pd.read_csv(os.path.join(DATA_PATH, "diner/diner_df_20241204_yamyam.csv"))

review = pd.DataFrame()
for review_data_path in review_data_paths:
    review = pd.concat([review, pd.read_csv(review_data_path)], axis=0)

review["reviewer_review_cnt"] = review["reviewer_review_cnt"].apply(lambda x: np.int32(str(x).replace(",", "")))
review = pd.merge(review, diner, on="diner_idx", how="inner")
```

```python
review.head()
```

```python
review = review.dropna(subset=["reviewer_review"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

stopwords = ["이", "그", "저", "그리고", "하지만", "은", "는", "이", "가", "을", "를"]

vectorizer = TfidfVectorizer(max_features=10, stop_words=stopwords)
tfidf_matrix = vectorizer.fit_transform(review["reviewer_review"])
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=[f"tfidf_{i}" for i in range(10)])
```
