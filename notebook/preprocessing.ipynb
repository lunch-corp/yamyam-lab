```python
# 특수 문자 제거하고 줄바꿈 문자를 공백으로 대체
import re
def remove_special_characters(text):
    return re.sub(r'[^\w\s]', '', text).replace('\n', ' ')

# review_df의 reviewer_review 컬럼 값 변경
review_df['reviewer_review'] = review_df['reviewer_review'].apply(remove_special_characters)
```

```python
# 형태소 분석
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt
```

```python
okt = Okt()

def custom_tokenizer(text):
    return okt.morphs(text)
```

```python
# 형태소로 토큰화
from tqdm import tqdm
tqdm.pandas()
review_df['tokenized_review'] = review_df['reviewer_review'].progress_apply(custom_tokenizer)
```

```python
# 불용어
stop_words = pd.read_csv('/Users/ranny/Downloads/stopword.txt', header=None)
stop_words.head()
```

```python
# 필요한 단어는 불용어에서 제거
not_stop_list = ['좋아', '아이고', '퉤', '에휴', '그저', '그럼']
stop_words = stop_words[~stop_words[0].isin(not_stop_list)]
```

```python
# 불용어만 리스트로 변환
stop_words = stop_words[0].to_list()
```

```python
# 불용어 제거 
def remove_stopwords(words):
    words = [word for word in words if word not in stop_words]
    return words
```

```python
tqdm.pandas()
review_df['without_stopwords'] = review_df['tokenized_review'].progress_apply(remove_stopwords)
```

```python
# 문자열로 변환
review_df['notstopwords'] = review_df['without_stopwords'].apply(lambda x : ' '.join(x)) 
```

```python
# 빈 문자열 삭제
review_df = review_df[review_df['notstopwords'] != '']
```

```python
# TF-IDF 행렬 생성
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(review_df['notstopwords'])
```

```python
# 행렬 데이터프레임으로 변환
tfidf_matrix_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())    
```

```python
import koreanize_matplotlib
```

```python
# 단어 구름 시각화

from wordcloud import WordCloud
import matplotlib.pyplot as plt 

font_path = "/System/Library/Fonts/Supplemental/AppleGothic.ttf" 

# 전체 TF-IDF 가중치의 합계를 구하여 단어 구름 생성
sum_tfidf = tfidf_matrix_df.sum(axis=0)
wordcloud = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate_from_frequencies(sum_tfidf)     

# 단어 구름 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud from TF-IDF")
plt.show()
```
